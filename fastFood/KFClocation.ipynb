```python
import requests
import json
KFC_locations = []
KFC_address = []
response = requests.get(
    'https://restapi.amap.com/v3/place/text?key=9013e06e0781b4f712e29a674d2c0d2f&types=050301&city=%E8%A5%BF%E5%AE%89&output=json').text
total_dict = json.loads(response)
for loc in total_dict['pois']:
    KFC_locations.append(loc['location'].split(","))
    KFC_address.append("KFC")
print(KFC_locations)
```

```python
McDonalds_locations = []
McDonalds_address = []
response = requests.get(
    "https://restapi.amap.com/v3/place/text?key=9013e06e0781b4f712e29a674d2c0d2f&types=050302&city=%E8%A5%BF%E5%AE%89&output=json").text
total_dict = json.loads(response)
for loc in total_dict['pois']:
    McDonalds_locations.append(loc['location'].split(","))
    McDonalds_address.append('McDonalds')
print(McDonalds_locations)
```

```python
Pizza_Hut_locations = []
Pizza_Hut_address = []
response = requests.get(
     "https://restapi.amap.com/v3/place/text?key=9013e06e0781b4f712e29a674d2c0d2f&types=050303&city=%E8%A5%BF%E5%AE%89&output=json").text
total_dict = json.loads(response)
for loc in total_dict['pois']:
    Pizza_Hut_locations.append(loc['location'].split(","))
    Pizza_Hut_address.append('Pizza_Hut_address')
print(Pizza_Hut_locations)
print(Pizza_Hut_address)
```

```python
ordinary_locations = []
ordinary_address = []
response = requests.get(
    "https://restapi.amap.com/v3/place/text?key=9013e06e0781b4f712e29a674d2c0d2f&keywords=||%E9%BA%A6%E4%B9%90%E9%A6%99||%E5%B8%8C%E5%A4%9A%E7%BE%8E%E7%82%B8%E9%B8%A1%E6%B1%89%E5%A0%A1||%E5%8D%8E%E8%8E%B1%E5%A3%AB&city=%E8%A5%BF%E5%AE%89&output=json").text

total_dict = json.loads(response)

for loc in total_dict['pois']:
    ordinary_locations.append(loc['location'].split(","))
    
    ordinary_address.append('ordinary_brand')
print(ordinary_locations)
print(len(ordinary_locations))
```

```python
#红色KFC，绿色McDonald，蓝色必胜客
import folium
latitude,longitude = 34.288567, 108.963041
xiAnMap = folium.Map([latitude,longitude])

KFC = folium.map.FeatureGroup()
MC = folium.map.FeatureGroup()
Pizza = folium.map.FeatureGroup()
ordinary_brand = folium.map.FeatureGroup()

for log,lat in KFC_locations:
    KFC.add_child(
        folium.CircleMarker(
            [float(lat),float(log)],
            radius=7,
            color='yellow',
              fill=True,
            fill_color='red',
            fill_opacity=1)   
    )

for log,lat in McDonalds_locations:
    MC.add_child(
        folium.CircleMarker(
            [float(lat),float(log)],
            radius=7,
            color='yellow',
            fill=True,
            fill_color='green',
            fill_opacity=1)
    )       
for log,lat in ordinary_locations:
    ordinary_brand.add_child(
    folium.CircleMarker(
    [float(lat),float(log)],
    radius=7,
        color="yellow",
        fill=True,
        fill_color = "black",
        fill_opacity = 1
    ))
# Pizza_Hut = folium.map.FeatureGroup()
# for log,lat in Pizza_Hut_locations:
#     Pizza_Hut.add_child(
#         folium.CircleMarker(
#             [float(lat),float(log)],
#             radius=7, 
#             color='yellow',
#             fill=True,
#             fill_color='blue',
#             fill_opacity=1
#         )
#     )
xiAnMap.add_child(KFC)
xiAnMap.add_child(MC)
# xiAnMap.add_child(Pizza_Hut)
xiAnMap.add_child(ordinary_brand)
xiAnMap.save("西安肯德基、麦当劳和一些小众品牌分布")
xiAnMap
```
