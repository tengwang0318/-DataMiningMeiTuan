```python
import folium
import requests
import json
response = requests.get("https://restapi.amap.com/v3/place/text?key=9013e06e0781b4f712e29a674d2c0d2f&keywords=%E5%85%B0%E5%B7%9E%E6%8B%89%E9%9D%A2&city=%E8%A5%BF%E5%AE%89&output=json").text
lanZhou = json.loads(response)
lanZhou_locations = []
for loc in lanZhou['pois']:
    lanZhou_locations.append(loc['location'].split(","))
print(lanZhou_locations)
```

```python
response = requests.get("https://restapi.amap.com/v3/place/text?key=9013e06e0781b4f712e29a674d2c0d2f&keywords=%E6%B2%99%E5%8E%BF%E5%B0%8F%E5%90%83&city=%E8%A5%BF%E5%AE%89&output=json").text
shaXian = json.loads(response)
shaXian_locations = []
for loc in shaXian['pois']:
    shaXian_locations.append(loc['location'].split(","))
print(shaXian_locations)
```

```python
latitude,longitude = 34.288567, 108.963041
xiAnMap = folium.Map([latitude,longitude])
shaXian = folium.map.FeatureGroup()
lanZhou = folium.map.FeatureGroup()

for log,lat in shaXian_locations:
    shaXian.add_child(
        folium.CircleMarker(
            [float(lat),float(log)],
            radius=7,
            color='yellow',
              fill=True,
            fill_color='red',
            fill_opacity=1)   
    )

for log,lat in lanZhou_locations:
    lanZhou.add_child(
        folium.CircleMarker(
            [float(lat),float(log)],
            radius=7,
            color='yellow',
              fill=True,
            fill_color='blue',
            fill_opacity=1)   
    )
xiAnMap.add_child(shaXian)
xiAnMap.add_child(lanZhou)
```
