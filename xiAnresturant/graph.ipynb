```python
import folium
import pandas as pd
latitude,longitude = 34.288567, 108.963041
xiAnMap = folium.Map([latitude,longitude])
xiAnMap.save("西安地图.html")
xiAnMap
```

```python
cdata = pd.read_csv("finalData.csv")

#西安市全部的店家的分布情况
incidents = folium.map.FeatureGroup()
for latitude,longitude in zip(cdata['latitude'],cdata['longitude']):
    incidents.add_child(
        folium.CircleMarker(
            [latitude, longitude],
            radius=7, 
            color='yellow',
            fill=True,
            fill_color='red',
            fill_opacity=0.4)
    )
    
xiAnMap.add_child(incidents)
xiAnMap.save("西安所有美团店家的分布情况.html")
```

```python
#西安市不同类别的店家的分布情况
latitude,longitude = 34.288567, 108.963041
xiAnMap = folium.Map([latitude,longitude])
best = folium.map.FeatureGroup()
medium = folium.map.FeatureGroup()
worst = folium.map.FeatureGroup()
for i in range(len(cdata['cluster'])):
    #最好
    if cdata['cluster'][i] == 3:
        best.add_child(
        folium.CircleMarker(
            [cdata['latitude'][i], cdata['longitude'][i]],
            radius=1, 
            color='red',
            fill=True,
            fill_color='red',
            fill_opacity=0.4)
    )
    #中等
    elif cdata['cluster'][i] == 1:
        medium.add_child(
        folium.CircleMarker(
            [cdata['latitude'][i], cdata['longitude'][i]],
            radius=1, 
            color='green',
            fill=True,
            fill_color='red',
            fill_opacity=0.4)
    )
    else :
        worst.add_child(
        folium.CircleMarker(
            [cdata['latitude'][i], cdata['longitude'][i]],
            radius=1, 
            color='blue',
            fill=True,
            fill_color='red',
            fill_opacity=0.4)
    )
xiAnMap.add_child(best)
xiAnMap.add_child(worst)
xiAnMap.add_child(medium)
xiAnMap.save("西安好中差三类店家的分布情况.html")
```

```python
#全部
latitudes,longitudes = [],[]
labels = []
for i in range(len(cdata['cluster'])):
    if cdata['cluster'][i]==3:
        latitudes.append(cdata['latitude'][i])
        longitudes.append(cdata['longitude'][i])
        labels.append('best')
    elif cdata['cluster'][i] == 2:
        latitudes.append(cdata['latitude'][i])
        longitudes.append(cdata['longitude'][i])
        labels.append('worst')
    else:
        latitudes.append(cdata['latitude'][i])
        longitudes.append(cdata['longitude'][i])
        labels.append('medium')

for lat, lng, label in zip(latitudes, longitudes, labels):
    folium.Marker([lat, lng], popup=label).add_to(xiAnMap)   
xiAnMap.save("西安所有店家的分类标签情况.html")
xiAnMap
```

```python
#所有等级的店家
from folium import plugins

latitude,longitude = 34.288567, 108.963041
xiAnMap = folium.Map([latitude,longitude])
latitudes,longitudes = [],[]
labels = []
for i in range(len(cdata['cluster'])):
    if cdata['cluster'][i]==3:
        latitudes.append(cdata['latitude'][i])
        longitudes.append(cdata['longitude'][i])
        labels.append('best')
    elif cdata['cluster'][i] == 2:
        latitudes.append(cdata['latitude'][i])
        longitudes.append(cdata['longitude'][i])
        labels.append('worst')
    else:
        latitudes.append(cdata['latitude'][i])
        longitudes.append(cdata['longitude'][i])
        labels.append('medium')

incidents = plugins.MarkerCluster().add_to(xiAnMap)


for lat, lng, label, in zip(latitudes,longitudes,labels):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# add incidents to map
xiAnMap.add_child(incidents)
xiAnMap.save("西安美团店家各区域的分布情况及数量特征.html")
```
